```python
#Lazar Novakovic
#Mandelbrot set homework.

#If you wish to see a zoomed point I found an interesting point you could use.

#increase points close to 1000

#Set domain to 0.00125

#Set i in y = width* (i-points/2) as (i-10.7*points)

#Set j in (x = width * (j-points/2)) as (j+227.5*points)

#Set k in elif k==100: as k==500:


#This is importing necessary tools for the Mandelbrot Set.

%matplotlib inline  
from numba import jit
from matplotlib.pyplot import *
import matplotlib.pyplot as plt
from math import *
from numpy import empty
import time
#This is defining values for the plotting process.


domain = 4 #This is the domain and range of the real and imaginary values used for the calculation.

points = 1000  #This dictates the amount of individual pixels will be calculated for the give graph, the resolution.

width = domain/points  #This is the distance between each point calculated.


div = empty([points,points],float) # This stores the values of the divergence of points from the set. Used for density plot. 
@jit
def brot():

    domain = 4 #This is the domain and range of the real and imaginary values used for the calculation.

    points = 5000  #This dictates the amount of individual pixels will be calculated for the give graph, the resolution.

    width = domain/points  #This is the distance between each point calculated.


    div = empty([points,points],float) # This stores the values of the divergence of points from the set. Used for density plot. 
    for i in range(0, points): #This runs through each of the y values, imaginary values.

        y = width* (i-points/2)    #This adjusts the distance between the points, resolution.

        # This also adjusts the center of the y.

        for j in range(0, points): #This runs through each of the real values.

            z = 0 #This resets the value of z to 0 for each new complex value.

            x = width * (j-points/2) #This adjusts the distance between the points, resolution.

            #This also adjusts the center of the x.

            k = 0 #This will reset the value of the divergence plotted.

            while True: #Here the iteration of each complex value will occur. 

                z = z**2 + complex(x,y)

                k = k + 1 #This will help count how many iterations have been done.

                if abs(z) >= 2: #If the value becomes larger than or equal to 2 the point is not in the set. 

                    div[i,j] = k #This will store that the point is out of the set.

                    # k was originally 1, which did not differentiate between and values of out of the set.
                    #Setting this to the amount of iterations before existing the set helps distinguish each point's divergence.

                    break

                elif k==250: #Once a point hasnt left the set after many iterations it will be labelled inside of the set.

                    div[i,j] = 0 #This gives the divergence 0, coloring it black.

                    break

    figure(figsize = (25,20)) #This increases the size of the plot in order to make it easier to see.

    #This is the density plot and the color type used.


    plt.imshow(div,origin='lower', extent = [-domain/2,domain/2,-domain/2,domain/2], aspect='auto',cmap="nipy_spectral") #This is the density plot

    plt.xlabel('Real', size = 20) #Label the Real axis.

    plt.ylabel('Imaginary', size=20) #Label the Imaginary axis.

    plt.colorbar() #Explains the colors of the plot.

    plt.show() #Shows the plot.

#I recognized that 2 significant shapes which point many points in the Mandelbrot set are a cardiod and the adjacent circle.
#Telling the code to by default color this black
#and ignore calculations of points within those 2 shape could also increase the calculation speed.
start_time = time.time()
brot()
print("--- %s seconds ---" % (time.time() - start_time))
```
