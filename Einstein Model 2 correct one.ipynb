```python
import matplotlib.pyplot as plt

from sympy import *

def Omega():
    
    n = input("What is N? ")

    Q = input("What is q? ")

    n = int(n)

    Q = int(Q)

    if 0 < n <= 3 and Q >= 0:
        for k in range(0, Q):
            ax = plt.subplots()

            plt.axis('off')

            ax = plt.axes([0,0,2.2,0.5]) 

            ax.set_xticks([])

            ax.set_yticks([])
            #y = factorial(Q + k - 1)/factorial(Q) * factorial(k -1)
            a = latex("\Omega") + latex(("= ")) + latex(factorial(k + n - 1)/factorial(k) * factorial(n -1))
            plt.text(0.03,(0.7),'$%s$' %a,size=21)
    else:
        print("N must be between 0 and 3, and q must be larger than 0 ")
```
