```python
#Lazar Novakovic
#Online Data fitting and transforms Homework

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from numpy import genfromtxt

#Similar to Lec_04 provided

data = np.loadtxt("circular.txt",float)

print(np.shape(data))

fdata = np.fft.fft2(data)

fdata[np.abs(fdata)<50] = 0

invdata = np.fft.ifft2(fdata)

plt.title('Data')

plt.imshow(data)

plt.colorbar()

plt.show()

plt.title('Transform Data')

plt.imshow(np.abs(fdata))

plt.colorbar()

plt.show()

plt.title('Inverse Transform')

plt.imshow(np.real(invdata))

plt.colorbar()

plt.show()
#plt.savefig(dat)
df = pd.DataFrame(invdata)

df.to_csv("circular.csv")
#Comparison of compression
```

```python
from numpy import genfromtxt
tesla = genfromtxt('TeslamultiTimeline.csv', delimiter=',')


#52 months of data, along with respective searches in an array.

x = len(tesla)

values = np.arange(x)

#Fits of the data
#Linear
lin = np.polyfit(values,tesla,1,full=True)

y1 = np.polyval(lin[0],values)
#Quadratic
quad = np.polyfit(values,tesla,2,full=True)

y2 = np.polyval(quad[0],values)
#Cubic
cube = np.polyfit(values,tesla,3,full=True)

y3 = np.polyval(cube[0],values)
#Quartic
quart = np.polyfit(values,tesla,4,full=True)

y4 = np.polyval(quart[0],values)
#Quintic
quint = np.polyfit(values,tesla,5,full=True)

y5 = np.polyval(quint[0],values)

#Plotting grids,labels, and fits
fig = plt.figure()
ax = fig.gca()
ax.set_xticks(np.arange(0, x, 5))
ax.set_yticks(np.arange(0, 100., 5))
plt.plot(values,y1,label= "linear")
plt.plot(values,y2,label= "quadratic")
plt.plot(values,y3,label= "cubic")
plt.plot(values,y4,label= "quartic")
plt.plot(values,y5,label= "quintic")
plt.grid()
plt.title('Fits of Tesla search data')
plt.xlabel('Months')
plt.ylabel('% of highest search value')
plt.plot
plt.legend(loc='upper left')
plt.show()
```

```python
#Tesla search data from Google Trends.
#The values provided by google are all percentages of the highest search value.


tesla = genfromtxt('TeslamultiTimeline.csv', delimiter=',')

x1 = len(tesla)
ivalues = np.arange(x1)

#The Transforms

Ttesla = np.fft.fft(tesla)
cutTtesla = Ttesla[np.abs(Ttesla)>5]
invtesla = np.fft.ifft(cutTtesla)

x2 = len(cutTtesla)
fvalues = np.arange(x2)

fig = plt.figure()
ax = fig.gca()
ax.set_xticks(np.arange(0, x1, 5))
ax.set_yticks(np.arange(0, 100., 5))

plt.plot(ivalues, tesla,'-', color = 'r')
plt.title('Data of Tesla Search Fit')
plt.xlabel('Time in Months')
plt.ylabel('Percentage of Max searched Value')
plt.grid()
plt.show()


plt.plot(fvalues,np.abs(cutTtesla), color = 'r')
plt.title('Data of Tesla Fit Transform')
plt.xlabel('Frequency Domain')
plt.ylabel('Amplitude')
plt.grid()
plt.show()

fig = plt.figure()
ax = fig.gca()
ax.set_xticks(np.arange(0, x1, 5))
ax.set_yticks(np.arange(0, 100., 5))

plt.plot(fvalues,np.abs(invtesla), color = 'r')
plt.title('Inverse Transform of Tesla Data')
plt.xlabel('Time in Months')
plt.ylabel('Percentage of Max searched value')
plt.grid()
plt.show()
```
