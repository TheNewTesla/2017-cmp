```python
#Lazar Novakovic Hw One Rule Algorithm 
from sklearn.datasets import load_iris
import numpy as np
from collections import OrderedDict
import pandas as pd
dataset = load_iris()

X = dataset.data
Y = dataset.target
print(X)
print(Y)

# In order to apply dimensionality reduction we can create a binary collection of the values, where the values above the mean are given a value of 1 and all others are 0.

X_d0 = np.array( X[:,0] > X[:,0].mean(), dtype = 'int')
X_d1 = np.array( X[:,1] > X[:,1].mean(), dtype = 'int')
X_d2 = np.array( X[:,2] > X[:,2].mean(), dtype = 'int')
X_d3 = np.array( X[:,3] > X[:,3].mean(), dtype = 'int')
#Reshaping the arrays

X_d0r = X_d0.reshape(150,1)
X_d1r = X_d1.reshape(150,1)
X_d2r = X_d2.reshape(150,1)
X_d3r = X_d3.reshape(150,1)
```

```python
#Load data as shown in lecture.
#Here the values will be binarily collected.
Iris_data_d = OrderedDict({'Sepal Length ': X_d0r[:,0],
             'Sepal Width': X_d1r[:,0],
             'Petal Length': X_d2r[:,0],
             'Petal Width': X_d3r[:,0],
             'Label': Y})

df = pd.DataFrame(Iris_data_d)
df
```

```python
Y_d = Y.reshape(150,1)
Z_d = np.hstack([X_d0r,X_d1r,X_d2r,X_d3r,Y_d])

a = ['Sepal length','Sepal width','Petal length','Petal width']
b = ['Setosa','Versicolor','Virginica']
c = '%'
results = []
for i in range(4):
    for j in range(3):
        selection = np.logical_and(Z_d[:,i]>.9,Z_d[:,4] == j)
        results.append(( a[i], b[j],'Results', len(Z_d[selection]),round((len(Z_d[selection])/len(Z_d[:,i]))*100)))
        #This is out of 50 samples each
```
